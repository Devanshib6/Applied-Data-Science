```python
from scipy import stats
```

```python
df['Request_Closing_Time']
```

```python
df['Response Time in Days'] = df['Request_Closing_Time'].dt.days
```

```python
df['Response Time in Days']
```

```python
df = df[df['Response Time in Days'] >= 0]
```

```python
grouped_data = df.groupby('Complaint Type')['Response Time in Days'].apply(list)
```

```python
f_statistic, p_value = stats.f_oneway(*grouped_data)

print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")
```

```python
alpha = 0.05

if p_value < alpha:
    print(f"\nResult: p-value = {p_value:.4f} < alpha = {alpha}")
    print("Conclusion: Reject the Null Hypothesis (H0).")
    print("There is statistically significant evidence that the average response time differs across at least one complaint type.")
else:
    print(f"\nResult: p-value = {p_value:.4f} >= alpha = {alpha}")
    print("Conclusion: Fail to reject the Null Hypothesis (H0).")
    print("There is not enough statistically significant evidence to conclude that the average response times differ across complaint types.")
```

```python
from scipy.stats import chi2_contingency
```

```python
df['Complaint Type']
```

```python
top_complaints = df['Complaint Type'].value_counts().head(5).index
df_top = df[df['Complaint Type'].isin(top_complaints)]
```

```python
# Drop rows with missing boroughs
df_top = df_top.dropna(subset=['Borough'])

# Step 3: Create contingency table
contingency_table = pd.crosstab(df_top['Complaint Type'], df_top['Borough'])

# Show the table (optional)
contingency_table
```

```python
from scipy.stats import chi2_contingency
# Perform Chi-Square Test
chi2, p_value, dof, expected = chi2_contingency(contingency_table)

# Show test result
print(f"Chi-Square Statistic: {chi2}")
print(f"P-value: {p_value}")
```

```python
alpha = 0.05  # 5% significance level

if p_value < alpha:
    print(f"The p-value ({p_value:.3f}) is less than the significance level ({alpha}).")
    print("Conclusion: We reject the Null Hypothesis (H0). There is statistically significant evidence to suggest that the average response time varies across different complaint types.")
else:
    print(f"The p-value ({p_value:.3f}) is greater than or equal to the significance level ({alpha}).")
    print("Conclusion: We fail to reject the Null Hypothesis (H0). There is not enough statistically significant evidence to conclude that the average response time differs significantly across complaint types.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from matplotlib.collections import LineCollection
df = df.copy()
df['Month'] = df['Created Date'].dt.month
monthly_trend = df['Month'].value_counts().sort_index()
x = np.array(monthly_trend.index)
y = np.array(monthly_trend.values)
points = np.array([x, y]).T.reshape(-1, 1, 2)
segments = np.concatenate([points[:-1], points[1:]], axis=1)
norm = plt.Normalize(y.min(), y.max())
lc = LineCollection(segments, cmap='coolwarm', norm=norm)
lc.set_array(y)
lc.set_linewidth(3)
fig, ax = plt.subplots(figsize=(12, 7))
ax.add_collection(lc)
ax.scatter(x, y, color='black', zorder=3)
all_month_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
                    'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
month_labels_filtered = [all_month_labels[i - 1] for i in x]
ax.set_xticks(x)
ax.set_xticklabels(month_labels_filtered, fontsize=12)
ax.set_ylabel('Number of Complaints', fontsize=14)
ax.set_xlabel('Month', fontsize=14)
ax.set_title('Monthly Complaint Volume Trend', fontsize=18, weight='bold')
max_idx = y.argmax()
min_idx = y.argmin()
ax.annotate(f'Peak: {y[max_idx]}', 
            xy=(x[max_idx], y[max_idx]), 
            xytext=(x[max_idx], y[max_idx] + 100),
            ha='center', fontsize=12, 
            arrowprops=dict(facecolor='green', shrink=0.05))
ax.annotate(f'Low: {y[min_idx]}', 
            xy=(x[min_idx], y[min_idx]), 
            xytext=(x[min_idx], y[min_idx] - 150),
            ha='center', fontsize=12, 
            arrowprops=dict(facecolor='red', shrink=0.05))
cbar = fig.colorbar(lc, ax=ax, orientation='vertical', label='Complaint Volume')
ax.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig('monthly_trend_decorated_gradient.png', dpi=300)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid", font_scale=1.1)
print("Data type of Request_Closing_Time:", df['Request_Closing_Time'].dtype)
print("Sample values:\n", df['Request_Closing_Time'].head(3))

try:
    df['Hours'] = df['Request_Closing_Time'].dt.total_seconds() / 3600
except AttributeError:
    df['Hours'] = df['Request_Closing_Time'] / 3600
avg_time = df.pivot_table(
    index='Complaint Type',
    columns='Borough',
    values='Hours',
    aggfunc='mean'
)
avg_time_long = avg_time.reset_index().melt(id_vars='Complaint Type', 
                                            var_name='Borough', 
                                            value_name='Average Hours')
boroughs = avg_time_long['Borough'].unique()
colors = sns.color_palette("Set2", len(boroughs))
for i, borough in enumerate(boroughs):
    plt.figure(figsize=(12, 8))
    borough_data = avg_time_long[avg_time_long['Borough'] == borough]
    ax = sns.barplot(
        data=borough_data,
        x='Average Hours',
        y='Complaint Type',
        color=colors[i],
        errorbar=None
    )
    plt.title(f'Average Request Closing Time in {borough}', fontsize=16)
    plt.xlabel('Average Hours to Close', fontsize=14)
    plt.ylabel('Complaint Type', fontsize=14)
    plt.tight_layout()
    for p in ax.patches:
        ax.annotate(f'{p.get_width():.2f}',
                    (p.get_width() + 0.1, p.get_y() + p.get_height() / 2),
                    ha='left', va='center', fontsize=12, color='black')
    plt.savefig(f"Average_Closing_Time_{borough}.png", format='png', dpi=300)
    plt.show()
```
